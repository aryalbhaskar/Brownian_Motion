## Practice Problem: Brownian Motion

Brownian motion is the random movement or fluctuation of particles in a medium. The process was discovered by a botanist (Robert Brown) while looking at grains of pollen moving in water. Later on this process was the basis for for discovering molecules and atoms. 

Brownian motion is sometimes called a random walk. To internalize this problem in two dimensions (i.e., x-y plane) imagine a particle at an initial position `x = 0` and `y = 0`. Then the particle moves some random (but short) distance in the `x` and `y` directions to reach a new position. Then the process repeats again from the new position. Thus, as time progresses, the particle drifts randomly on the two-dimensional plane.

```python
# Brownian Motion / Random Walk 

# import library 
import numpy as np
import matplotlib.pyplot as plt

# initial particle positon in x and y aixis 
x = 0
y = 0

# Empty list to append update value of locations 
x_pos = []
y_pos = []

# Define loop and iterations, 2000 itereatins in this case
Walk = 2000
# Set seed for reproducibility 
np.random.seed(5)
for walk in range(Walk):
    x = x + np.random.uniform(-1, 1) # update x-location with unifrom random movement 
    x_pos.append(x)
    y = y + np.random.uniform(-1, 1) # update y-location with unifrom random movement
    y_pos.append(y)

# Plot the random path 
plt.figure(figsize=(8,6))
plt.title('Random Walk')
plt.plot(x_pos, y_pos, '-+',  linewidth=0.25)
plt.scatter(0, 0, marker = 'o', facecolor = 'r')
plt.scatter(x_pos[Walk-1], y_pos[Walk-1], marker = 'o', facecolor = 'r')
plt.xlabel('X', size=14)
plt.ylabel('Y', size=14)
```
